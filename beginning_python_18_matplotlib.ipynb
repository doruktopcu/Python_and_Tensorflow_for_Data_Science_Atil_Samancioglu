```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  
#for ease of use
```

```python
age_list = [10, 20, 30, 40, 50, 60]
weight_list = [35, 105, 67, 60, 70, 60]
```

```python
numpy_age_list = np.array(age_list)
numpy_weight_list = np.array(weight_list)
```

```python
plt.plot(numpy_age_list, numpy_weight_list, "b") 
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show() # no need because of %matplotlib inline
```

```python
numpy_array_1 = np.linspace(0,10,20)
```

```python
numpy_array_1
```

```python
numpy_array_2 = numpy_array_1 ** 2
```

```python
plt.plot(numpy_array_1, numpy_array_2, "g*-") 
# * + - can be used
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpy_array_1, numpy_array_2, "r+-")
plt.subplot(1,2,2)
plt.plot(numpy_array_2, numpy_array_1, "g*-")
```

```python
my_figure = plt.figure()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.2, 0.2, 0.7, 0.7])
figure_axes.plot(numpy_array_1, numpy_array_2, "g")
figure_axes.set_xlabel("Ages")
figure_axes.set_ylabel("Weights")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
figure_2 = plt.figure()

axis_1 = figure_2.add_axes([0.1, 0.1, 0.9, 0.9])
axis_2 = figure_2.add_axes([0.2, 0.5, 0.3, 0.3])

axis_1.plot(numpy_array_1, numpy_array_2, "b")
axis_1.set_xlabel("X_1")
axis_1.set_ylabel("Y_1")
axis_1.set_title("Main Plot Title 1")

axis_2.plot(numpy_array_2, numpy_array_1, "g")
axis_2.set_xlabel("X_2")
axis_2.set_ylabel("Y_2")
axis_2.set_title("Sub Plot Title 2")

```

```python
(my_figure, my_axes) = plt.subplots()
my_axes.plot(numpy_array_1, numpy_array_2, "b")
```

## Subplots  

```python
(my_figure, my_axes) = plt.subplots(nrows = 1, ncols = 2)

for axis in my_axes:
    axis.plot(numpy_array_1, numpy_array_2, "g")
    axis.set_xlabel("X axis")
plt.tight_layout()
```

```python
new_figure = plt.figure("my_figure.png", dpi = 300)

new_plot = new_figure.add_axes([0.1, 0.1, 0.9, 0.9])
new_plot.plot(numpy_array_1, numpy_array_1 **2, "*-", label = "numpyArray ** 2")
new_plot.plot(numpy_array_1, numpy_array_1 **3, "*-", label = "numpyArray ** 3")
new_plot.legend()
```

```python
new_figure.savefig("my_figure.png", dpi = 300)
```

## Matplotlib Visualization

```python
(my_new_figure, my_new_axes) = plt.subplots()
my_new_axes.plot(numpy_array_1, numpy_array_2 + 25 , color = "#269DE6", alpha = 0.5, linewidth = 2.5, linestyle = ":")
my_new_axes.plot(numpy_array_2, numpy_array_1 + 20, color = "#4126E6", alpha = 0.8, linewidth = 2, linestyle = "--")
my_new_axes.plot(numpy_array_2, numpy_array_1 + 15, color = "#E68D26", alpha = 0.8, linewidth = 1, linestyle = "-")
my_new_axes.plot(numpy_array_2, numpy_array_1 + 10, color = "#E626AA", alpha = 0.8, linewidth = 1.5, linestyle = "-.")
my_new_axes.plot(numpy_array_2, numpy_array_1 + 0, color = "#E626AA", alpha = 0.8, linewidth = 1.5, linestyle = "--", marker = "o", markersize = 5)
```
